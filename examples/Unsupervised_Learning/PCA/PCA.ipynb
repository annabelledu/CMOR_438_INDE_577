## Principal Component Analysis (PCA)

Principal Component Analysis is a **dimensionality reduction technique** that
projects data onto orthogonal directions of maximum variance.

PCA is commonly used for:
- Visualization
- Noise reduction
- Feature compression

In this notebook, we:
- Standardize data
- Apply PCA
- Visualize the low-dimensional projection


```python
import numpy as np
import matplotlib.pyplot as plt

from rice_ml import train_test_split, standardize
```

```python
from rice_ml.unsupervised_learning.pca import PCA

rng = np.random.default_rng(0)
X = rng.normal(0, 1, size=(300, 3))
X[:,2] = X[:,0]*0.8 + rng.normal(0, 0.2, size=300)  # correlated feature

X_s, params = standardize(X, return_params=True)
pca = PCA(n_components=2).fit(X_s)
Z = pca.transform(X_s)

print("Explained variance ratio:", pca.explained_variance_ratio_)

plt.scatter(Z[:,0], Z[:,1], s=10)
plt.title("PCA projection (2D)")
plt.show()
```
